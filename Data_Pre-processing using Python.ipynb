```python
#Performing Normalisation
# Create a copy of the dataframe
df_normalized = df.copy()

# List of numeric columns to scale
numeric_cols = ['person_age', 'person_income', 'loan_amnt', 'loan_int_rate']

# Manual Min-Max Scaling formula: (X - min) / (max - min)
for col in numeric_cols:
    df_normalized[col] = (df_normalized[col] - df_normalized[col].min()) / (df_normalized[col].max() - df_normalized[col].min())

# View the result
df_normalized[numeric_cols].head()
```

```python
#Performing Standardisation
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
standard_scaler = StandardScaler()

# Create a copy of the dataframe
df_standardized = df.copy()

# Apply Standardization to key numeric columns
df_standardized[numeric_cols] = standard_scaler.fit_transform(df_standardized[numeric_cols])

# View the result (values will scale around 0)
df_standardized[numeric_cols].head()
```

```python
#Identifying and Removing Outliers
# 1. Inspect extreme values before filtering
print("Max Age before:", df['person_age'].max())
print("Max Employment Length before:", df['person_emp_length'].max())

# 2. Filter out unrealistic rows
df = df[(df['person_age'] <= 100) & (df['person_emp_length'] <= 60)]

# 3. Confirm clean boundaries
print("\nMax Age after:", df['person_age'].max())
print("Max Employment Length after:", df['person_emp_length'].max())
```

```python
#Checking Data Formats
# List of text/categorical columns
text_cols = ['person_home_ownership', 'loan_intent', 'cb_person_default_on_file']

# Clean white spaces and force uppercase formatting
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.upper()

# Verify uniform formatting
print(df['person_home_ownership'].unique())
```

```python
#Grouping Ages
# Define the age boundary brackets and their corresponding business labels
age_bins = [0, 25, 35, 50, 100]
age_labels = ['Young Adult (<25)', 'Core Professional (25-35)', 'Mid-Career (35-50)', 'Senior (50+)']

# Create the engineered categorical column
df['age_group'] = pd.cut(df['person_age'], bins=age_bins, labels=age_labels, right=False)

# Check the distribution of your new column
print(df['age_group'].value_counts())
```

```python
#Grouping Incomes
# Define income thresholds based on common portfolio tiers
# (0 to 35k, 35k to 70k, 70k to 100k, and 100k+)
income_bins = [0, 35000, 70000, 100000, float('inf')]
income_labels = ['Low Income (<35k)', 'Medium Income (35k-70k)', 'High Income (70k-100k)', 'HNW (>100k)']

# Create the engineered income slab column
df['income_slab'] = pd.cut(df['person_income'], bins=income_bins, labels=income_labels, right=False)

# Check the distribution of income segments
print(df['income_slab'].value_counts())
```

```python
#Exporting clean dataset 
# Save the final cleaned dataframe to a new CSV file
df.to_csv(r"C:\Shambhavi\Finance Project Credit Risk Assessment\Dataset\Cleaned_Credit_Risk_Dataset.csv", index=False)
print("Dataset cleaned, engineered, and successfully exported!")
```
